## 1. DuckDB のインポートと接続

`duckdb.connect()` でDB接続を作ります。  
引数なし → **インメモリDB**（プロセス終了でデータ消滅）  
引数あり → **ファイルDB**（`.duckdb` ファイルに永続化）

```python
import duckdb
import os

# バージョン確認
print(f"DuckDB version: {duckdb.__version__}")
```

```python
# インメモリDB（引数なし）
con = duckdb.connect()
print("インメモリDBに接続しました")
```

---
## 2. CSV をそのまま SQL でクエリ（DuckDBの最大の特徴）

DuckDB は **CSVファイルをテーブルとして直接クエリできます**。  
データベースにインポートする手間が不要！

```python
# CSVファイルのパス（このノートブックから見た相対パス）
DATA_DIR = "../data"

# CSVを直接SELECTする
result = con.execute(f"""
    SELECT *
    FROM read_csv_auto('{DATA_DIR}/customers.csv')
""").fetchdf()  # fetchdf() でpandas DataFrameとして取得

print("=== customers.csv の内容 ===")
result
```

```python
# products.csv も同様に
con.execute(f"""
    SELECT *
    FROM read_csv_auto('{DATA_DIR}/products.csv')
""").fetchdf()
```

```python
# sales.csv の最初の5件だけ確認
con.execute(f"""
    SELECT *
    FROM read_csv_auto('{DATA_DIR}/sales.csv')
    LIMIT 5
""").fetchdf()
```

---
## 3. インメモリDB vs ファイルDB

DuckDB のDBは `.duckdb` という**1ファイル**が実体です。

```python
# ファイルDBを作成
DB_PATH = "../mydb.duckdb"

file_con = duckdb.connect(DB_PATH)
print(f"ファイルDBを作成しました: {DB_PATH}")

# ファイルが実際に存在するか確認
if os.path.exists(DB_PATH):
    size_kb = os.path.getsize(DB_PATH) / 1024
    print(f"ファイルサイズ: {size_kb:.1f} KB")
```

```python
# ファイルDBにテーブルを作成してみる
file_con.execute("""
    CREATE TABLE IF NOT EXISTS test_table (id INTEGER, value TEXT)
""")
file_con.execute("INSERT INTO test_table VALUES (1, 'hello'), (2, 'world')")

# データ確認後、接続を閉じる
print(file_con.execute("SELECT * FROM test_table").fetchdf())
file_con.close()

# 再接続してもデータが残っていることを確認
file_con2 = duckdb.connect(DB_PATH)
print("\n再接続後も残っているか？")
print(file_con2.execute("SELECT * FROM test_table").fetchdf())
file_con2.close()
```

---
## 4. テーブル作成・データ投入・SELECT

CSVを読み込んでDuckDB内にテーブルとして取り込みます。

```python
# CSVからテーブルを作成
con.execute(f"""
    CREATE TABLE customers AS
    SELECT * FROM read_csv_auto('{DATA_DIR}/customers.csv')
""")

con.execute(f"""
    CREATE TABLE products AS
    SELECT * FROM read_csv_auto('{DATA_DIR}/products.csv')
""")

con.execute(f"""
    CREATE TABLE sales AS
    SELECT * FROM read_csv_auto('{DATA_DIR}/sales.csv')
""")

print("3テーブルを作成しました")

# テーブル一覧の確認
con.execute("SHOW TABLES").fetchdf()
```

```python
# テーブルのスキーマ（カラム情報）確認
con.execute("DESCRIBE sales").fetchdf()
```

```python
# 基本SELECT
con.execute("""
    SELECT *
    FROM sales
    WHERE amount > 100000
    ORDER BY amount DESC
    LIMIT 10
""").fetchdf()
```

---
## 5. 集計クエリ

DuckDB は分析クエリが得意です。

```python
# 月別売上集計
con.execute("""
    SELECT
        strftime(sale_date, '%Y-%m') AS year_month,
        COUNT(*)                     AS 件数,
        SUM(amount)                  AS 売上合計,
        AVG(amount)                  AS 平均単価
    FROM sales
    GROUP BY year_month
    ORDER BY year_month
""").fetchdf()
```

```python
# カテゴリ別売上（productsテーブルと組み合わせ）
con.execute("""
    SELECT
        p.category,
        COUNT(*)       AS 件数,
        SUM(s.amount)  AS 売上合計
    FROM sales s
    JOIN products p ON s.product_id = p.product_id
    GROUP BY p.category
    ORDER BY 売上合計 DESC
""").fetchdf()
```

---
## 6. JOIN クエリ

3テーブルを結合して、人間が読みやすいレポートを作ります。

```python
# 売上明細（顧客名・商品名付き）
con.execute("""
    SELECT
        s.sale_id,
        s.sale_date,
        c.name        AS 顧客名,
        c.region      AS 地域,
        p.product_name AS 商品名,
        p.category    AS カテゴリ,
        s.quantity    AS 数量,
        s.amount      AS 金額
    FROM sales s
    JOIN customers c ON s.customer_id = c.customer_id
    JOIN products  p ON s.product_id  = p.product_id
    ORDER BY s.sale_date
    LIMIT 10
""").fetchdf()
```

```python
# 顧客別売上ランキング
con.execute("""
    SELECT
        c.name       AS 顧客名,
        c.region     AS 地域,
        COUNT(*)     AS 購入回数,
        SUM(s.amount) AS 累計購入額
    FROM sales s
    JOIN customers c ON s.customer_id = c.customer_id
    GROUP BY c.customer_id, c.name, c.region
    ORDER BY 累計購入額 DESC
""").fetchdf()
```

---
## まとめ

| 機能 | 説明 |
|------|------|
| `read_csv_auto()` | CSVをそのままSQLでクエリ（インポート不要） |
| `duckdb.connect()` | インメモリDB（一時的） |
| `duckdb.connect('path.duckdb')` | ファイルDB（永続化） |
| `CREATE TABLE AS SELECT` | CSVからテーブルを作成 |
| `DESCRIBE テーブル名` | スキーマ確認 |
| `SHOW TABLES` | テーブル一覧 |
| `.fetchdf()` | 結果をpandas DataFrameで取得 |
